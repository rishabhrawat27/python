# Importing Library

```python
import pandas as pd
```

# Importing Data

```python
df1=pd.read_excel(r"C:\Users\iamri\Downloads\Customer Call List.xlsx")
df1
```

# Droping Duplicate and Not useful Columns

```python
df1.drop(columns='Not_Useful_Column',inplace=True)
df1= df1.drop_duplicates()
df1
```

# Using Strip Function

```python
#df1['Last_Name']=df1['Last_Name'].str.lstrip('...')
#df1['Last_Name']=df1['Last_Name'].str.lstrip('/')
#df1['Last_Name']=df1['Last_Name'].str.rstrip('_')
df1['Last_Name']=df1['Last_Name'].str.strip('_./')
df1
```

# Removing Unwanted from Phone number

```python
df1['Phone_Number']=df1['Phone_Number'].str.replace('[^a-zA-Z0-9]','')
df1
```

# Converting Phone Number to string

```python
df1['Phone_Number']=df1['Phone_Number'].apply(lambda x: str(x))
```

# Formating The Phone Number

```python
df1['Phone_Number']=df1['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df1['Phone_Number']=df1['Phone_Number'].str.replace('--','')
df1
```

# Formating House Column

```python
df1[['House_No.','House_Address']]=df1['Address'].str.split(' ',1,expand=True)
df1
```

```python
df1[["Street_Name","City","Postal_Code"]]=df1["House_Address"].str.split(',',2,expand=True)
df1
```

```python
df1 = df1.drop(columns=["Address","House_Address"])
df1
```

# Removing nan and Na From Phone Number and Last Name

```python
df1['Phone_Number']=df1['Phone_Number'].str.replace('nan','')
df1['Phone_Number']=df1['Phone_Number'].str.replace('Na','')
df1['Last_Name']= df1['Last_Name'].apply(lambda x : str(x))
df1['Last_Name']= df1['Last_Name'].str.replace("NaN","")
df1['Last_Name']= df1['Last_Name'].str.replace("nan","")
df1
```

# Adding First and Last name

```python
df1["First_Name"]=df1["First_Name"]+' '+df1["Last_Name"]
df1
```

```python
df1=df1.drop(columns='Last_Name')
df1
```

```python
#df1["Postal_Code"]=df1["Postal_Code"].str.replace('None',' ')
#df1
```

# Formating City

```python
df1["Postal_Code"]=df1["Postal_Code"].apply(lambda x : str(x))
df1["City"]=df1["City"].apply(lambda x : str(x))
```

```python
df1['City']=df1['City'].str.replace('None',' ')
df1
```

# Formating Paying Customer and Do not Contact

```python
df1['Paying Customer'] = df1['Paying Customer'].str.replace("Yes","Y")
df1['Paying Customer'] = df1['Paying Customer'].str.replace("No","N")
df1
```

```python
df1['Do_Not_Contact'] = df1['Do_Not_Contact'].str.replace("Yes","Y")
df1['Do_Not_Contact'] = df1['Do_Not_Contact'].str.replace("No","N")
df1
```

```python
df1['Paying Customer']=df1['Paying Customer'].str.replace("N/a",'')
df1
```

# Droping Empty Phone Number Columns

```python
for x in df1.index :
    if df1.loc[x,"Phone_Number"] == "":
        df1.drop(x,inplace=True)

df1
```

# Droping Y in Do not Contact

```python
for x in df1.index :
    if df1.loc[x,"Do_Not_Contact"] == 'Y':
        df1.drop(x, inplace=True)
```

# Reindexing colunmns 

```python
df1=df1.reindex(columns=['CustomerID','First_Name','Phone_Number','House_No.','Street_Name','City','Paying Customer','Do_Not_Contact'])
df1
```

```python
df1.rename(columns={'First_Name':'Full_Name'},inplace=True)
df1
```
