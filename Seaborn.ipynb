# Importing Libraries

```python
import os
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

# Importing Data

```python
df1 = pd.read_csv(r"C:\Users\iamri\OneDrive\Desktop\car data all together.csv")
df1
```

# Ploting Price Of Cars Using matplotlib

```python
plt.plot(df1["price"])
plt.xticks(range(0,len(df1["price"])),df1["CarName"],rotation=90)
plt.show()
```

# Ploting Price Of Cars Using Seaborn

```python
sns.set(rc={"figure.figsize":(14,10)})
ax=sns.lineplot(x="CarName",y="price",data=df1)
ax.tick_params(axis="x",rotation=90)
```

# Ploting Different Sedan cars Price, Grouped by cylinder number and aspiration.

-The average price of Turbo Cars in the same cylinder category is more then the Standard one.

-As the number of cylinder in the Sedan cars increases the price also increases.

```python
Sedan = df1[df1["carbody"]=="sedan"].reset_index(drop=True)
sns.set(rc={"figure.figsize":(14,10)})
ax=sns.lineplot(x="CarName",y="price",hue="cylindernumber",style="aspiration",data=Sedan)
ax.tick_params(axis="x",rotation=90)
```

# Mean Price of Different Carbody grouped by FuelType

-The Heighest mean price if of Hardtop diesel cars (28176) and lowest is of Hatchback diesel Cars (7788)

-There is no Diesel cars in 'Convertable' cars

```python
sns.set(rc={"figure.figsize":(14,10)})
ax=sns.barplot(x='carbody',y='price',hue="fueltype",data=df1,ci=False)

for i in ax.containers:
    ax.bar_label(i)

df1[['carbody',"fueltype",'price']].groupby(['carbody',"fueltype"],as_index=False).mean()
```

# Finding Relationship of price and different aspects, Groupby fueltype

```python
categories=['aspiration','doornumber','drivewheel','enginelocation','fuelsystem']

for i in categories:
    ax=sns.barplot(x=i,y='price',hue="fueltype",data=df1,ci=False)
    for i in ax.containers:
        ax.bar_label(i)
    plt.show()
```

# Histogram of price showing mean price line

```python
ax=sns.distplot(df1['price'])
plt.axvline(df1['price'].mean())
```

# Extracting Numercial Columns of the dataset.

```python
num = ["int64","float64"]
var = df1.select_dtypes(include =num).columns
df5 = df1[var]
df5
```

# Histogram of all the numerical columns

```python
for i in df5:
    a = df5[i].values
    sns.distplot(a)
    plt.title(i)
    plt.show()
```

# Boxplot and Swarmplot, grouped by Fueltype

-Four , Five and Six cylender car have outliers

-Average price of diesel car is more then gas one

```python
sns.boxplot(x='cylindernumber',y='price',hue='fueltype',data=df1)
sns.swarmplot(x='cylindernumber',y='price',hue='fueltype',data=df1)
```

# Scatterplot showing relationship between wheelbase and price, Grouped by Carbody, Horsepower and fueltype

```python
sns.scatterplot(x='wheelbase',y='price',data=df1,hue='carbody',size='horsepower',style='fueltype',s=27)
```

# Regression line between Price and horsepower

```python
sns.lmplot(x='price',y='horsepower',data=df1)
```

# Different regression line showing relationship between price and horsepower, Grouped by carbody

```python
sns.lmplot(x='price',y='horsepower',data=df1,hue='carbody',markers=['.','+','*',',','_'],height=6,ci=False)
```

 # Different regression plot showing relationship between price and horsepower, Grouped by drivewheel

```python
sns.lmplot(x='price',y='horsepower',hue='drivewheel',data=df1,col='carbody',col_wrap=2,height=6,ci=False)
```

# Subplots

```python

fig, axes= plt.subplots(2,2,figsize=(15,10))

#plot 1
sns.boxplot(x="price",y="carbody",data=df1, ax=axes[0,0],color="#c50000")
sns.swarmplot(x="price",y="carbody",data=df1, ax=axes[0,0],color="#f19524")
axes[0,0].set_title("price of different carbodies")

sns.distplot(df1['horsepower'],ax=axes[0,1],color="#c50000")
axes[0,1].set_title("horsepower frequency")

sns.histplot(df1['enginesize'],ax=axes[1,0],color="#c50000")
axes[1,0].set_title("enginesize frequency")

sns.boxplot(x="horsepower",y="carbody",data=df1,ax=axes[1,1],color="#c50000")
sns.swarmplot(x="horsepower",y="carbody",data=df1,ax=axes[1,1],color="#f19524")
axes[1,1].set_title("horsepower of different carbodies")


plt.show()
```
